```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/home/blu-bridge005/Desktop/Anuj@BluBridge/TensorParallel/CustomDNN/gpt2_tests/training_logs/customdnn_training_log9.csv')

plt.plot(df['step'].to_numpy(), df['loss'].to_numpy())

plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.savefig('loss_curve.png')

print("Maximum throughput: ", df['tok_per_sec'].max())
print("Avg. Throughput:",df['tok_per_sec'].mean())
print("Min Train Loss:",df['loss'].min())
print("Average loss: ", df['loss'].mean())
print("Max Grad Norm: ", df['grad_norm'].max())
print("Min Val Loss:",df['val_loss'].where(df['val_loss']!=(-1)).min())
```
