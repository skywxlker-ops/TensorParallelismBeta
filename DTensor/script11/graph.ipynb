```python
import numpy as np
import os

def convert_npy_to_bin(npy_filename, bin_filename):
    """
    Converts a .npy file (which has a NumPy header) to a raw binary .bin file 
    (without a header).

    Args:
        npy_filename (str): The path to the input .npy file.
        bin_filename (str): The path to the output .bin file.
    """
    try:
        # 1. Load the data from the .npy file into a NumPy array
        # np.load automatically handles the .npy header
        data = np.load(npy_filename)
        print(f"Successfully loaded data from {npy_filename} with shape: {data.shape} and dtype: {data.dtype}")

        # 2. Write the raw data to the .bin file using tofile()
        # This method writes the array data as a raw stream of bytes
        with open(bin_filename, 'wb') as f:
            data.tofile(f)
        
        print(f"Successfully wrote raw binary data to {bin_filename}")

    except FileNotFoundError:
        print(f"Error: The file {npy_filename} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- Example Usage ---
if __name__ == "__main__":
    # 1. Create a sample .npy file first for demonstration
    # # sample_data = np.array([1.5, 2.7, 3.14, 4.0], dtype=np.float32)
    # sample_npy_filename = 'sample_data.npy'
    # np.save(sample_npy_filename, sample_data)
    # print(f"Created sample file: {sample_npy_filename}")

    # 2. Convert the created .npy file to a .bin file
    output_bin_filename = 'val_shard_00000.bin'
    convert_npy_to_bin("/home/blubridge-035/Desktop/Backup/parallelism/script11/val_shard_00000.npy", output_bin_filename)

    # Optional: Verify the output by reading the .bin file back
    if os.path.exists(output_bin_filename):
        # We must specify the exact dtype and shape to read raw binary correctly
        # In this case, float32 and shape of (4,)
        with open(output_bin_filename, 'rb') as f:
            loaded_data = np.fromfile(f, dtype=np.float32)
        print(f"\nVerification: Loaded data from {output_bin_filename} as: {loaded_data}")
        # print(f"Verification successful: {np.array_equal(sample_data, loaded_data)}")
        
        # Clean up the sample files
        # os.remove(sample_npy_filename)
        # os.remove(output_bin_filename)
        # print("Cleaned up sample files.")
```
