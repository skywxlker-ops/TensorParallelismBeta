```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('DTensor/TP_MLP_Training_logs/TP_MLP_Training_log25.csv')
df2 = pd.read_csv('/home/blu-bridge005/Desktop/Anuj@BluBridge/TensorParallel/CustomDNN/gpt2_tests/Naive/naive_run_log6.csv')
# df3 = pd.read_csv('/home/blu-bridge005/Desktop/Anuj@BluBridge/TensorParallel/CustomDNN/gpt2_tests/training_logs/customdnn_training_log18.csv')

plt.figure(figsize=(10, 6))
plt.plot(df1['step'].to_numpy(), df1['loss'].to_numpy(), label='DNN - with admm kernel', color='blue', alpha=0.8)
plt.plot(df2['step'].to_numpy(), df2['loss'].to_numpy(), label='CustomDNN - with admm kernel', color='red', alpha=0.5)
# plt.plot(df3['step'].to_numpy(), df3['loss'].to_numpy(), label='without any sync', color='green', alpha=1)

print("Maximum throughput (DNN - with admm kernel): ", df1['tok_per_sec'].max())
print("Avg. Throughput (DNN - with admm kernel):",df1['tok_per_sec'].mean())
print("Min Train Loss (DNN - with admm kernel):",df1['loss'].min())
print("Average loss (DNN - with admm kernel): ", df1['loss'].mean())
print("Max Grad Norm (DNN - with admm kernel): ", df1['grad_norm'].max())
print("Min Val Loss (DNN - with admm kernel):",df1['val_loss'].where(df1['val_loss']!=(-1)).min())

print("\n---------------------------------------------------------------------------------------------\n")

print("Maximum throughput (CustomDNN - with admm kernel): ", df2['tok_per_sec'].max())
print("Avg. Throughput (CustomDNN - with admm kernel):",df2['tok_per_sec'].mean())
print("Min Train Loss (CustomDNN - with admm kernel):",df2['loss'].min())
print("Average loss (CustomDNN - with admm kernel): ", df2['loss'].mean())
print("Max Grad Norm (CustomDNN - with admm kernel): ", df2['grad_norm'].max())
print("Min Val Loss (CustomDNN - with admm kernel):",df2['val_loss'].where(df2['val_loss']!=(-1)).min())

plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.grid(True)
# plt.savefig('combined_loss_curve.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('DTensor/TP_MLP_Training_logs/TP_MLP_Training_log25.csv')
df2 = pd.read_csv('/home/blu-bridge005/Desktop/Anuj@BluBridge/TensorParallel/CustomDNN/gpt2_tests/Naive/naive_run_log6.csv')

plt.figure(figsize=(10, 6))
plt.plot(df1['step'].to_numpy(), df1['tok_per_sec'].to_numpy(), label='DNN - with admm kernel', color='blue', alpha=0.8)
plt.plot(df2['step'].to_numpy(), df2['tok_per_sec'].to_numpy(), label='CusomDNN - with admm kernel', color='red', alpha=0.8)

print("max throughput - DNN - with admm kernel = ", df1['tok_per_sec'].max())
print("max throughput - CustomDNN - with admm kernel = ", df2['tok_per_sec'].max())

print("average throughput - DNN - with admm kernel = ", df1['tok_per_sec'].mean())
print("average throughput - CustomDNN - with admm kernel = ", df2['tok_per_sec'].mean())

print("std dev - DNN - with admm kernel = ", df1['tok_per_sec'].std())
print("std dev - CustomDNN - with admm kernel = ", df2['tok_per_sec'].std())

plt.xlabel('Step')
plt.ylabel('tok_per_sec')
plt.title('Training tok_per_sec Comparison')
plt.legend()
plt.grid(True)
# plt.savefig('combined_loss_curve.png')
plt.show()
```
